Главная задача, которой мы будем заниматься, - алгоритмы исследования систем алгебраических уравнений.
"Алгебраических" означает, что эти уравнения заданы многочленами.

Важные частные случаи этой задачи вам должны быть известны.

В случае, если все уравнения линейны, такая система решается методом Гаусса.

В другом случае, когда неизвестная переменная ровно одна, задачу можно свести к решению одного уравнения от одной неизвестной.
Таким уравнением будет НОД левых частей исходных уравнений.

На этом занятии мы вспомним алгоритм Евклида для вычисления НОД двух чисел.
Аккуратно написанный алгоритм Евклида можно будет применять и к многочленам от одной переменной.

Вот нерекурсивная версия алгоритма Евклида:

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a
```

Вот рекурсивная:

```python
def gcd(a, b):
    if b != 0:
        return gcd(b, a % b)
    return a
```

Обратите внимание, что в этой реализации мне нигде не пришлось сравнивать два элемента с помощью < или >.
Я выполняю сравнения только на равенство нулю.

Задание: написать теперь расширенный алгоритм Евклида.
Расширенный алгоритм кроме самого НОД должен вернуть ещё два элемента u и v, такие, что НОД(a, b) = au + bv.

Напишите также тесты к вашему алгоритму, предусмотрев особые случаи (когда один из аргументов - ноль, или когда аргументы отрицательные и т. д.)
